```python
import gspread
import pandas as pd
from google.oauth2.service_account import Credentials
from dotenv import load_dotenv
import os
import re

load_dotenv()
```

```python
SERVICE_ACCOUNT = os.getenv("SERVICE_ACCOUNT")
SCOPES = ['https://www.googleapis.com/auth/spreadsheets', 'https://www.googleapis.com/auth/drive']


creds = Credentials.from_service_account_file(SERVICE_ACCOUNT, scopes=SCOPES)
client = gspread.Client(creds)

ss = client.open_by_url("https://docs.google.com/spreadsheets/d/1LlKOei13aiJ018ZQxvXj3hwz1ESDm133Er6stsxk76c/edit?gid=1039699707#gid=1039699707",)
ws = ss.worksheet("Leads Masterlist")
```

```python
emails = ws.col_values(11)
emails
```

```python
REGEX = r'@([A-Za-z0-9.-]+\.[A-Za-z]{2,})'
domains = []

for item in emails:
    match = re.search(REGEX, item)
    if match:
        domain = match.group(1)
        domains.append(domain)

domains
```

```python
df_domains = pd.DataFrame(domains, columns=["domains"])
df_domains
```

```python
df_domains.value_counts()
```

```python
df_unique_domains = df_domains.drop_duplicates(ignore_index=True)
df_unique_domains
```

```python
unique_domains = df_unique_domains['domains'].tolist()
```

```python
from dns.resolver import resolve

def is_valid_domain(domain: str) -> bool:
    try:
        resolve(domain, 'MX')
        return True
    except:
        try:
            resolve(domain, 'A')
            return True
        except:
            return False
```

```python
accepted_domains = []

for item in unique_domains:
    if is_valid_domain(item):
        accepted_domains.append(item)
        
accepted_domains
```
